```python
# Cell 1: Get OHLC Input
print("üéØ DOJI IDENTIFIER")
print("=" * 50)

open_price = float(input("Enter OPEN price: "))
high_price = float(input("Enter HIGH price: "))
low_price = float(input("Enter LOW price: "))
close_price = float(input("Enter CLOSE price: "))

print(f"\nüìä Your Candle:")
print(f"   Open:  {open_price}")
print(f"   High:  {high_price}")
print(f"   Low:   {low_price}")
print(f"   Close: {close_price}")
```

```python
# Cell 2: Analyze for Doji
# Determine candle color
is_green = close_price > open_price
is_red = close_price < open_price

# Calculate wicks
if is_green:
    upper_wick = high_price - close_price
    lower_wick = open_price - low_price
    candle_color = 'GREEN üü¢'
else:
    upper_wick = high_price - open_price
    lower_wick = close_price - low_price
    candle_color = 'RED üî¥' if is_red else 'NEUTRAL ‚ö™'

body = abs(close_price - open_price)
price_range = high_price - low_price

# Calculate percentages
body_percent = (body / price_range) * 100 if price_range > 0 else 0
upper_wick_percent = (upper_wick / close_price) * 100 if close_price > 0 else 0
lower_wick_percent = (lower_wick / close_price) * 100 if close_price > 0 else 0

# Calculate wick difference
avg_wick = (upper_wick + lower_wick) / 2
wick_diff = abs(upper_wick - lower_wick)
wick_diff_percent = (wick_diff / avg_wick) * 100 if avg_wick > 0 else 0

print("\n" + "=" * 50)
print("üìè CANDLE MEASUREMENTS")
print("=" * 50)
print(f"Candle Color:      {candle_color}")
print(f"Upper Wick:        {upper_wick:.2f} ({upper_wick_percent:.2f}%)")
print(f"Lower Wick:        {lower_wick:.2f} ({lower_wick_percent:.2f}%)")
print(f"Body:              {body:.2f} ({body_percent:.2f}%)")
print(f"Wick Difference:   {wick_diff:.2f} ({wick_diff_percent:.2f}%)")

# Check doji criteria
WICK_DIFF_THRESHOLD = 30  # Wicks must be within 30% similarity
MIN_WICK_PERCENT = 1      # Each wick must be at least 1% of price
MAX_BODY_PERCENT = 30     # Body must be less than 30% of range

is_doji = (
    wick_diff_percent <= WICK_DIFF_THRESHOLD and
    upper_wick_percent >= MIN_WICK_PERCENT and
    lower_wick_percent >= MIN_WICK_PERCENT and
    body_percent < MAX_BODY_PERCENT
)

print("\n" + "=" * 50)
print("üîç DOJI CRITERIA CHECK")
print("=" * 50)
print(f"‚úì Wick similarity ({wick_diff_percent:.1f}% ‚â§ {WICK_DIFF_THRESHOLD}%):  {'PASS ‚úÖ' if wick_diff_percent <= WICK_DIFF_THRESHOLD else 'FAIL ‚ùå'}")
print(f"‚úì Upper wick > {MIN_WICK_PERCENT}% ({upper_wick_percent:.1f}%):         {'PASS ‚úÖ' if upper_wick_percent >= MIN_WICK_PERCENT else 'FAIL ‚ùå'}")
print(f"‚úì Lower wick > {MIN_WICK_PERCENT}% ({lower_wick_percent:.1f}%):         {'PASS ‚úÖ' if lower_wick_percent >= MIN_WICK_PERCENT else 'FAIL ‚ùå'}")
print(f"‚úì Small body < {MAX_BODY_PERCENT}% ({body_percent:.1f}%):         {'PASS ‚úÖ' if body_percent < MAX_BODY_PERCENT else 'FAIL ‚ùå'}")

print("\n" + "=" * 50)
if is_doji:
    # Determine doji type
    if upper_wick > body * 2 and lower_wick < body:
        doji_type = 'DRAGONFLY DOJI üêâ'
        signal = 'BULLISH REVERSAL üìà'
        description = 'Long lower wick, short upper wick - buyers pushing back'
    elif lower_wick > body * 2 and upper_wick < body:
        doji_type = 'GRAVESTONE DOJI ü™¶'
        signal = 'BEARISH REVERSAL üìâ'
        description = 'Long upper wick, short lower wick - sellers pushing back'
    elif upper_wick > body * 2 and lower_wick > body * 2:
        doji_type = 'LONG-LEGGED DOJI ü¶µ'
        signal = 'HIGH VOLATILITY ‚ö°'
        description = 'Long wicks on both sides - extreme indecision'
    else:
        doji_type = 'STANDARD DOJI ‚≠ê'
        signal = 'INDECISION ü§î'
        description = 'Balanced wicks - market uncertainty'

    print("üéØ RESULT: THIS IS A DOJI CANDLE!")
    print("=" * 50)
    print(f"Type:        {doji_type}")
    print(f"Signal:      {signal}")
    print(f"Meaning:     {description}")
    print("\n‚ö†Ô∏è  WARNING: This candle pattern may indicate trend reversal!")
else:
    print("‚ùå RESULT: NOT A DOJI CANDLE")
    print("=" * 50)
    print("This is a regular candle pattern.")

print("\n" + "=" * 50)
```
